```python

# Step 1: Get all soil columns
soil_cols = [col for col in df.columns if col.startswith("Soil_Type")]

# Step 2: Convert one-hot → single column (number)
df["Soil_Type_Number"] = df[soil_cols].idxmax(axis=1)

# Extract number from 'Soil_Type7' → 7
df["Soil_Type_Number"] = df["Soil_Type_Number"].str.extract(r'(\d+)').astype(int)

# Step 3: Mapping (number → actual name)
soil_mapping = {
    1: "Cathedral family - Rock outcrop complex",
    2: "Vanet - Ratake families complex",
    3: "Haploborolis - Rock outcrop complex",
    4: "Ratake family - Rock outcrop complex",
    5: "Vanet family - Rock outcrop complex",
    6: "Vanet - Wetmore families complex",
    7: "Gothic family",
    8: "Supervisor - Limber families complex",
    9: "Troutville family",
    10: "Bullwark - Catamount families complex",
    11: "Bullwark - Catamount families complex (variant)",
    12: "Legault family",
    13: "Catamount family",
    14: "Pachic Argiborolis",
    15: "Unspecified",
    16: "Cryaquolis",
    17: "Gateview family",
    18: "Rogert family",
    19: "Typic Cryaquolis",
    20: "Typic Cryaquepts",
    21: "Typic Cryaquolis (variant)",
    22: "Leighcan family",
    23: "Leighcan family - Rock outcrop complex",
    24: "Leighcan family - Rock land complex",
    25: "Leighcan family - Rock land complex (variant)",
    26: "Leighcan family - Rock outcrop complex (variant)",
    27: "Leighcan family - Cryaquolis complex",
    28: "Leighcan family - Cryaquepts complex",
    29: "Leighcan family - Cryoborolis complex",
    30: "Cryoborolis - Rock outcrop complex",
    31: "Cryoborolis - Rock land complex",
    32: "Cryumbrepts - Rock outcrop complex",
    33: "Cryumbrepts - Rock land complex",
    34: "Cryumbrepts - Cryaquolis complex",
    35: "Cryumbrepts - Cryaquepts complex",
    36: "Bross family - Rock land complex",
    37: "Rock outcrop - Bross family complex",
    38: "Rock land - Bross family complex",
    39: "Cryumbrepts - Rock outcrop complex (variant)",
    40: "Cryumbrepts - Rock land complex (variant)"
}

# Step 4: Map numbers to names
Soil_Type_Name = df["Soil_Type_Number"].map(soil_mapping)



print("✅ Done! New columns added.")
```

```python
# Features & target
X = df.drop(["Cover_Type"], axis=1)
y = df["Cover_Type"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```
